```python
import pandas as pd
from IPython.display import Markdown, display
def printmd(string):
    display(Markdown(string))
    
import numpy as np

df = pd.read_csv("People.csv",float_precision='round_trip')
df['debut'] = df['debut'].astype('datetime64[ns]')
df['finalGame'] = df['finalGame'].astype('datetime64[ns]')
```
